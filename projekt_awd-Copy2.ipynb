```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns

plt.style.use('seaborn')
```

```python
dane = pd.read_csv("vgsales.csv")
dane
```

```python
grupy=[['1980-1989'],['1990-1999'],['2000-2009'],['2010-2020']]
gen = iter(range(4))
for i in range(1980,2020,10):
    grupy[next(gen)].append(sum(lata.Global_Sales.loc[i:i+9]))
grupy[3][1] += lata.Global_Sales.loc[2020]
grupy=dict(grupy)
plt.bar(grupy.keys(),grupy.values())
```

```python
lata_act=dane[dane.Genre=='Action'].groupby('Year').sum()
lata_plat=dane[dane.Genre=='Platform'].groupby('Year').sum()
lata_act
```

```python
act_dict=dict(lata_act.loc[:,'Global_Sales'])
plat_dict = dict(lata_plat.loc[:,'Global_Sales'])
plt.plot(act_dict.keys(),act_dict.values(),label='Action')
plt.plot(plat_dict.keys(),plat_dict.values(),label='Platform',color='orange')
#kiedy Year jest objektem:
# plt.xticks([0,len(act_dict.keys())//2,len(act_dict.keys())-1])
#kiedy Year jest intem
#plt.xticks(rotation=80)
plt.title('Sprzedaż gier akcji i platformowych w latach 1980-2017')
plt.xlabel('Rok [-]')
plt.ylabel('Liczba sprzedanych egzemplarzy [-]')
#plt.xlim([1979,2016])
plt.legend()
plt.show()
```

```python
t,p = st.ttest_ind(lata_act['Global_Sales'],lata_plat['Global_Sales'])
print('t =',round(t,5),'\np =',round(p,5))
print(lata_act['Global_Sales'].mean(),lata_plat['Global_Sales'].mean(),sep='\n')
```

```python
sims_dodatki = {'The Sims':9,'The Sims 2':11,'The Sims 3':10,'The Sims 4':3}
plt.bar(sims_sprzedaz.keys(),sims_sprzedaz.values())
plt.xlabel('Część gry [-]')
plt.ylabel('Sprzedane egzemplarze [mln]')
```

```python
plt.bar(sims_dodatki.keys(),sims_dodatki.values(),color='purple')
```

```python
st.pearsonr(list(sims_sprzedaz.values()),list(sims_dodatki.values()))
```

```python
plt.scatter(sims_dodatki.values(),sims_sprzedaz.values(),sizes=[50,50,50,50])
```

```python
d = np.polyfit(list(sims_dodatki.values()),list(sims_sprzedaz.values()),1)
f = np.poly1d(d)
plt.scatter(sims_dodatki.values(),sims_sprzedaz.values(),sizes=[50,50,50,50])
plt.xlabel('Liczba dodatków [-]')
plt.ylabel('Liczba sprzedanych egzemplarzy gry [mln]')
plt.plot(sims_dodatki.values(),f(list(sims_dodatki.values())),label='Linia regresji',color='r')
plt.legend()
plt.title('Zależność między liczbą dodatków do gry, a liczbą sprzedanych egzemplarzy podstawowej wersji gry')
```
